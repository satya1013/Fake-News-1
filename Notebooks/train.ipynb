```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../Dataset/train.csv", header=0, index_col=0)
df.head(10)
```

```python
df.isnull().sum(axis = 0)
```

```python
df['author'] = df['author'].fillna('prskid1000')
df.isnull().sum(axis = 0)
```

```python
df.dropna(subset = ["title", "text"], inplace=True) 
df.isnull().sum(axis = 0)
```

```python
len(df)
```

```python
#df['author'].unique()
#Calculating probility of an article written by an author is not fake
prob = {}
for i in df['author'].unique():
    t = len(df[(df['label'] == 1) & (df['author'] == i)])
    f = len(df[(df['label'] == 0) & (df['author'] == i)])
    prob[i] = t / (t + f)
print(prob)
```

```python
new_col = []
for i in df['author']:
    new_col.append(prob[i])
print(new_col)
```

```python
df['Author Honesty'] = new_col
df.head(10)
```

```python
df.to_csv('../Dataset/trainReady.csv', index=False)
```
